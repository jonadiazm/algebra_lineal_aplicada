```python
import numpy as np
import matplotlib.pyplot as plt
from utils import plot_matrix
```

```python
A = np.array([[3, 7], [5, 2]])
print(A)
```

```python
U, D, V = np.linalg.svd(A)
```

```python
sigma = np.zeros((U.shape[1], V.shape[0]))  # Matriz de ceros con el tama√±o correcto
np.fill_diagonal(sigma, D)
U @ sigma @ V
```

```python
plot_matrix(np.eye(2))
plot_matrix(V)
plt.show()

plot_matrix(np.eye(2))
plot_matrix(sigma)
plt.show()

plot_matrix(np.eye(2))
plot_matrix(U)
plt.show()

plot_matrix(np.eye(2))
plot_matrix(A)
plt.show()
```
