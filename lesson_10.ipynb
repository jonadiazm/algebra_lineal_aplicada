```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 1000)

y1 = -4*x + 3
y2 = 2*x + 5
y3 = -3*x + 1

plt.plot(x, y1, label='y=-4x+3')
plt.plot(x, y2, label='y=2x+5')
plt.plot(x, y3, label='y=-3x+1')

plt.xlim([-3, 2.5])
plt.ylim([-5, 9])
plt.show()
```

```python
matriz = np.array([[4, 1], [-2, 1], [3, 1]])
print(matriz)
```

```python
b = np.array([[3], [5], [1]])
b
```

```python
matriz_pinv = np.linalg.pinv(matriz)
```

```python
result = matriz_pinv @ b
```

```python
x = np.linspace(-5, 5, 1000)

y1 = -4*x + 3
y2 = 2*x + 5
y3 = -3*x + 1

plt.plot(x, y1, label='y=-4x+3')
plt.plot(x, y2, label='y=2x+5')
plt.plot(x, y3, label='y=-3x+1')

plt.scatter(result[0], result[1], color='red')

plt.xlim([-3, 2.5])
plt.ylim([-5, 9])
plt.show()
```
