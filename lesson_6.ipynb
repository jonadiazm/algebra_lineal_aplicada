```python
import numpy as np
import matplotlib.pyplot as plt
from utils import plot_matrix
```

```python
A = np.array([[3, 7], [5, 2]])
print("A:")
print(A)
```

```python
print("Transformación A:")
plot_matrix(np.eye(2))
plot_matrix(A)
```

```python
U, D, V = np.linalg.svd(A)

print("U:")
print(U)

print("D:")
print(D)

print("V:")
print(V)
```

```python
print("Primer rotación:")
plot_matrix(np.eye(2), alpha=0.4)
plot_matrix(V)
plt.show()

print("Escala:")
plot_matrix(np.eye(2), alpha=0.4)
plot_matrix(np.diag(D))
plt.show()

print("Segunda rotación:")
plot_matrix(np.eye(2), alpha=0.4)
plot_matrix(U)
plt.show()

print("Producto de las matrices:")
plot_matrix(np.eye(2), alpha=0.4)
plot_matrix(U @ np.diag(D) @ V)
plt.show()
```
