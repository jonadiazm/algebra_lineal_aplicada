```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [8, 4]

A = imread("images/stray_dog_513872.jpg")

X = np.mean(A, -1)  # Convert RGB to grayscale

channel = 1
A_f = np.zeros_like(A)
# A_f[:, :, 0] = A[:, :, 0]
# A_f[:, :, 1] = A[:, :, 1]
A_f[:, :, 2] = A[:, :, 2]

img = plt.imshow(X)
# img = plt.imshow(A[:, :, 2])
# img = plt.imshow(A_f)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
U, D, V = np.linalg.svd(X, full_matrices=False)
print(U.shape, D.shape, V.shape)
```

```python
imagen_recons = np.matrix(U[:, :1]) * np.diag(D[:1]) * np.matrix(V[:1, :])
plt.imshow(imagen_recons, cmap='gray')
plt.axis('off')
plt.show()
```

```python
i = 25
imagen_recons = np.matrix(U[:, :i]) * np.diag(D[:i]) * np.matrix(V[:i, :])
plt.imshow(imagen_recons, cmap="gray")
plt.axis("off")

title = f"Valor singular {i}"
plt.title(title)
plt.show()
```
