```python
import numpy as np
import matplotlib.pyplot as plt
from utils import plot_vectors
```

```python
np.random.randn(200).mean()
```

```python
np.random.seed(42)

x = 3 * np.random.rand(200)
y = 20 * x + 2 * np.random.randn(200)

x = x.reshape(200, 1)
y = y.reshape(200, 1)
```

```python
xy = np.hstack((x, y))
```

```python
plt.plot(xy[:, 0], xy[:, 1], '.')
plt.show()
```

```python
xy_centrado = xy - np.mean(xy, axis=0)
```

```python
plt.plot(xy_centrado[:, 0], xy_centrado[:, 1], '.')
plt.show()
```

```python
autovalores, autovectores = np.linalg.eig(xy_centrado.T @ xy_centrado)
```

```python
plot_vectors(autovectores.T, ["blue", "red"])
plt.plot(xy_centrado[:, 0], xy_centrado[:, 1], '.')
plt.show()
```

```python
xy_nuevo = (autovectores.T @ xy_centrado.T).T
```

```python
xy_nuevo[:, 0]
```

```python
plot_vectors(autovectores, ["blue", "red"])
plt.plot(xy_nuevo[:, 0], xy_nuevo[:, 1], '.')
plt.show()
```
